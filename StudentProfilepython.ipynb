```python
name = input("Enter your name")
age = int(input("Enter your age"))

if age >=18:
    age_status = "Adult"
else:
    age_status ="minor"
    
gpa = float(input("Enter your gpa"))

if gpa > 4.0:
    Deans_List = "Invalid gpa"
    
elif gpa>=3.5:
    Deans_List = "Yes"
else:
    Deans_List = "No"

    
courses = int(input("How many courses do you offer"))
string_list = []
for i in range (courses):
    s = input(f"courses # {i+1}:")
    string_list.append(s)
    
student_profile = {}
student_profile ["name"] = name
student_profile ["age"] = age
student_profile ["GPA"] = gpa
student_profile ["courses"] = courses


print("===============STUDENT PROFILE=============")
print (f"Name: {name}")
print (f"Age: {age}")
print (f"gpa: {gpa}")
print (f"courses {courses}")
print (f"Deans_List: {Deans_List}")
print (f"Age Status: {age_status}")
print (f"Num_courses: {int(courses)}")
print (f"Courses: {string_list}")
```
